## Apply Central Limit Theorem
Large sample sizes are mostly non-normal. Used CLT to convert the data to normal:
draw 50 random samples of size 30 and use the **sample means** as the working dataset.

```python
# storage of means
means_data = {"Solar": [], "Wind": []}

np.random.seed(42)

# sampling means
for _ in range(50):  # 50 times
    for c in ["Solar", "Wind"]:
        sample = np.random.choice(df[c], size=30, replace=True)
        means_data[c].append(np.mean(sample))
```

```python
df_means = pd.DataFrame(means_data)
df_means
```

```python
df_means.to_csv(r"dataset/electricity-clt-dataset.csv", index=False)
print("CLT dataset saved!")
```
