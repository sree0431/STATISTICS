```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'dataset\transformer-voltage-10-sample.csv',
                delimiter=",",
                index_col='Sample')
df 
```

```python
df.info()
```

```python
df.describe()
```

```python
alpha = 0.05
n = df['Voltage'].count()

t_critical = stats.t.ppf(1-alpha/2,n-1)
t_critical
```

```python
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

ci_lower, ci_upper = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = mean,
    scale = std/np.sqrt(n)
)

print(f"Confidence Interval = [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 300)
p = stats.norm.pdf(x, mean, std)

confidence_level = 0.95
z = stats.norm.ppf((1 + confidence_level) / 2)

ci_lower = mean - z * std
ci_upper = mean + z * std

plt.figure(figsize=(12, 6), facecolor='#F8F5F0')

plt.plot(
    x, p,
    color="#3A2C1C",
    linewidth=3.2,
    label="Normal Distribution",
    zorder=3
)

plt.hist(
    df['Voltage'],
    bins=7,
    density=True,
    alpha=0.75,
    color="#7D5C46",
    edgecolor="#3A2C1C",
    linewidth=1.4,
    label="Observed Data",
    zorder=2
)

plt.fill_between(
    x, p,
    where=(x >= ci_lower) & (x <= ci_upper),
    color="#3A2C1C",
    alpha=0.15,
    label=f"{confidence_level*100:.0f}% CI Area"
)

plt.axvline(
    x=ci_lower,
    color="#8B4513",
    linestyle="--",
    linewidth=2.2,
    label=f"Lower CI ({ci_lower:.2f}V)",
    zorder=4
)

plt.axvline(
    x=ci_upper,
    color="#A0522D",
    linestyle="--",
    linewidth=2.2,
    label=f"Upper CI ({ci_upper:.2f}V)",
    zorder=4
)

plt.axvline(
    x=mean,
    color="#2C1810",
    linestyle="-",
    linewidth=2,
    alpha=0.7,
    label=f"Mean ({mean:.2f}V)"
)

plt.title(
    "Voltage Distribution Analysis",
    font="Georgia",
    fontsize=14,
    pad=20,
    weight='bold'
)

plt.xlabel(
    "Voltage (V)",
    fontsize=11,
    labelpad=10
)

plt.ylabel(
    "Probability Density",
    fontsize=11,
    labelpad=10
)

plt.grid(
    True,
    alpha=0.3,
    linestyle=':',
    linewidth=0.8
)

plt.legend(
    frameon=True,
    framealpha=0.9,
    edgecolor="#3A2C1C",
    fontsize=10,
    loc='upper right'
)

plt.tight_layout()
plt.show()
```

```python
mean_size_6 = df.iloc[0,1:13].mean()
mean_size_6
```

```python
std_size_6 = df.iloc[0,1:13].mean()
std_size_6
```

```python
n = df.shape[1]-1 #returns 13 in (17,13)

lower_6, upper_6 = stats.t.interval(
confidence = 0.95,
df = n-1,
loc = mean_size_6,
scale = std_size_6/np.sqrt(n)
)

print(f"CI_6:[{lower_6: .2f}, {upper_6: .2f}]")
```

```python

mean = df.iloc[0:17,1:13].mean(axis = 1)
std = df.iloc[0:17,1:13].std(axis = 1)

df['ci_lower'], df['ci_upper'] = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = mean,
    scale = std/np.sqrt(n)
)

print(f"Confidence Interval = [{ci_lower:.2f}, {ci_upper:.2f}]")
```
