## Current Power

```python
#import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'dataset/current-power.csv',
                delimiter=",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
x = df['Current']
y = df['Power']

plt.figure(figsize=(10,6))
plt.scatter(x, y)
plt.title('Scatter plot of Current vs Power')
plt.xlabel('Current (A)')
plt.ylabel('Power (W)')
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(10,6))
plt.scatter(x, y, label='Data points')
plt.plot(x, line, color='red', label=f'Regression line (r_value={r_value:.3f})')
plt.title('Device Current vs Power Response')
plt.xlabel('Current (A)')
plt.ylabel('Power (W)')
plt.legend()
plt.show()
```

## Air Quality

### NOx

```python
air = pd.read_csv(r'dataset/air-quality.csv',
                  delimiter=",")
air
```

```python
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(10,6))
plt.scatter(x, y, color = "#f928", label='Data points')
plt.plot(x, line, color="black", alpha=0.2, label=f'Regression line (r = {r_value:.3f})')
plt.title('NOx vs Humidity Correlation')
plt.xlabel('NOx ug/cu.m')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

### NO2

```python
x = air['NO2']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(10,6))
plt.scatter(x, y, color = "violet", label='Data points')
plt.plot(x, line, color="black", alpha=0.2, label=f'Regression line (r = {r_value:.3f})')
plt.title('NO2 vs Humidity Correlation')
plt.xlabel('NO2 ug/cu.m')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

## Temperature - Humidity Correlation

```python
x = air['Temperature']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(10,6))
plt.scatter(x, y, color = "pink", label='Data points')
plt.plot(x, line, color="black", alpha=0.2, label=f'Regression line (r = {r_value:.3f})')
plt.title('Temperature vs Humidity Correlation')
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```
